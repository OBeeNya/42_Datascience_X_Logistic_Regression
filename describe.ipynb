```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('datasets/dataset_train.csv')
```

```python
data.drop('Index', axis=1,
          inplace=True)
```

```python
data_num = data.select_dtypes(include = np.number)
```

```python
describe = pd.DataFrame(index = range(8),
                        columns=range(len(data_num.columns)))
```

```python
describe.index = ['Count',
                  'Mean',
				  'Std',
                  'Min',
                  '25%',
                  '50%',
                  '75%',
                  'Max']
```

```python
describe.columns = data_num.columns
```

```python
for i in range(len(describe.columns)):
	# Count
	describe.iloc[0, i] = data_num.iloc[:, i].notnull().sum()
	# Mean
	describe.iloc[1, i] = np.sum(data_num.iloc[:, i]) / describe.iloc[0, i]
	# Standard deviation
	describe.iloc[2, i] = np.sqrt(np.sum((data_num.iloc[:, i] - describe.iloc[1, i]) ** 2) / (describe.iloc[0, i] - 1))
	# Minimum
	describe.iloc[3, i] = data_num.iloc[0, i]
	for j in range(len(data_num.iloc[:, i])):
		if data_num.iloc[j, i] < describe.iloc[3, i]:
			describe.iloc[3, i] = data_num.iloc[j, i]
	# First quartile
	describe.iloc[4, i] = data_num.iloc[:, i].sort_values().iloc[int((describe.iloc[0, i] + 1) / 4)]
	# Second quartile
	describe.iloc[5, i] = data_num.iloc[:, i].sort_values().iloc[int((describe.iloc[0, i] + 1) / 2)]
	# Third quartile
	describe.iloc[6, i] = data_num.iloc[:, i].sort_values().iloc[3 * int((describe.iloc[0, i] + 1) / 4)]
	# Maximum
	describe.iloc[7, i] = data_num.iloc[0, i]
	for j in range(len(data_num.iloc[:, i])):
		if data_num.iloc[j, i] > describe.iloc[7, i]:
			describe.iloc[7, i] = data_num.iloc[j, i]
```
